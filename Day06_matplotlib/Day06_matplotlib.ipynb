```python
#1. Importing and Setup
import matplotlib.pyplot as plt
import numpy as np

print("Imported Matplotlib and Numpy successfully")
```

```python
#2. Line Plot

x=np.arange(10)
y=np.sin(x)

plt.plot(x,y)
plt.title("Line plot of Sin(x)")
plt.xlabel("x")
plt.ylabel("Sin(x)")
plt.grid(True)
plt.show()
print("Line plot displayed: sin curve")
```

```python
#3. Bar Plot

categories=['Python','C++','Java']
students=[80,60,75]

plt.bar(categories,students, color='orange')
plt.title("Bar Plot of Language Popularity")
plt.ylabel("No of students")
plt.show()
print("Bar chart of programming languages displayed")
```

```python
#4. Scatter Plot

x=np.random.rand(50)
y=np.random.rand(50)

plt.scatter(x,y, color='green')
plt.title("Random scatter plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
#5. Histogram

data=np.random.randn(1000)

plt.hist(data,bins=30,color='purple')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
print("Histogram displayed for normal distribution")
```

```python
#6. Subplots

x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.subplot(1,2,1)
plt.plot(x,y1,'b--')
plt.title("Sin")

plt.subplot(1,2,2)
plt.plot(x,y2,'r--')
plt.title("Cosine")

plt.tight_layout()
plt.savefig("plot_name.png")
plt.show()
print("Two subplots displayed: sin and cos")
```

```python
#7. Customization Options

x=np.linspace(0,10,100)
y=np.sin(x)

plt.plot(x,y,linestyle='-.',color='red',label='sin(x)')
plt.title("Customized Plot")
plt.xlabel("Angle in rad")
plt.ylabel("sin value")

plt.legend()
plt.grid(True)
plt.show()
print("Customized plot with labels, grid and legend")
```

```python
#Challenge 1: Horizontal Bar Plot
#Make a horizontal bar plot showing favorite sports of 5 people.

sports=["Cricket","Football","Chees","Tennis","Badminton"]
votes=[100,70,80,60,85]

plt.barh(sports,votes,color='teal')
plt.title("favorite sports")
plt.xlabel("Number of votes")
plt.ylabel("Sports")
plt.show()
```

```python
#Challenge 2: Multiple Line Plots on Same Graph
#lot sin(x), cos(x), and tan(x) on the same graph with legends.

x = np.linspace(-2*np.pi, 2*np.pi, 500)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)

plt.plot(x,y1,label="Sin(x)",color='blue')
plt.plot(x,y2,label="Cin(x)",color='green')
plt.plot(x,y3,label="Tan(x)",color='red')

plt.ylim(-10,10)
plt.title("Trigonometric Functions")
plt.xlabel("x (radians)")
plt.ylabel("y values")
plt.axhline(0,color='black',linewidth=0.5)
plt.axvline(0,color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Challenge 3: Scatter Plot with Colors
#Create a scatter plot with 100 random points, where color varies with value.

np.random.seed(0)
x=np.random.rand(100)
y=np.random.rand(100)
colors=x+y

plt.scatter(x,y,c=colors,cmap='viridis',s=100,edgecolors='black')
plt.colorbar(label='x+y')
plt.title("Colored Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```
